Data disini dibaca langsung dari file yang ada pada google drive yakni data bernama: data-penambahan-kasus-covid-19-bulan-mei-tahun-2020

```
covid_sum = covid.groupby(['tanggal','diagnosa'])['jumlah'].sum()
covid_sum
```

Pada bagian ini untuk melakukan groupby, disini saya groupby dengan tanggal dan diagnosa

```
ax=sns.scatterplot(x='jumlah', y='tanggal', hue='diagnosa', style='diagnosa', data=covid)
```

Untuk melakukan visualisasi menggunakan seaborn, disini saya buat scatterplot, terdapat 3 buah diagnosa yakni meninggal, sembuh dan positif. Sumbu x adalah jumlah, sumbu ya adalah tanggal.
